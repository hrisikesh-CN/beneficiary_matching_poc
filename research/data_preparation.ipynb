```python
import pandas as pd

df = data_1.copy()

# Step 1: Determine the most frequent (Longitude, Latitude) pair for each Address
# with fallback to the first available pair if mode is empty
def get_most_frequent_or_first(group):
    mode_result = group[['longitude', 'latitude']].mode()
    if not mode_result.empty:
        return mode_result.iloc[0]
    else:
        return group[['longitude', 'latitude']].iloc[0]  # Fallback to first row if mode is empty

correct_mapping = (
    df.groupby('Address')
    .apply(get_most_frequent_or_first)
    .reset_index()
)

# Convert the correct mapping into a dictionary for easy lookup
mapping_dict = correct_mapping.set_index('Address')[['longitude', 'latitude']].to_dict('index')

# Step 2: Function to update longitude and latitude values in the DataFrame based on the correct mapping
def update_coordinates(row):
    correct_long, correct_lat = mapping_dict[row['Address']]['longitude'], mapping_dict[row['Address']]['latitude']
    row['longitude'] = correct_long
    row['latitude'] = correct_lat
    return row

# Step 3: Apply the function to modify the original DataFrame
df = df.apply(update_coordinates, axis=1)

# Display the modified DataFrame
print("Modified DataFrame with Consistent Longitude and Latitude:")
print(df)
```

```python
import numpy as np
from sklearn.metrics import pairwise_distances

def haversine(u, v):
    lon1, lat1 = u[0], u[1]
    lon2, lat2 = v[0], v[1]
    
    # Convert degrees to radians
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    
    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    r = 6371  # Radius of Earth in kilometers
    return c * r
```

```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'latitude': [40.7128, 34.0522, 51.5074],
    'longitude': [-74.0060, -118.2437, -0.1278]
})

# Apply the haversine distance function
distances = pairwise_distances(df[['longitude', 'latitude']], metric=haversine)

print(distances)
```
